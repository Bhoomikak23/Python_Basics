```python
#sample dataframe
import pandas as pd
data_frame = pd.DataFrame({
    'roll_no' : [102,103,104,105,101],
    'name' : ['bhoomi','venu','dg','mona','ranju'],
    'grade' : ['A','B','A','C','B'],
    'marks' : [15,15,20,4,22],
    'city' : ['dvg','sokke','tghalli','hubli','jgl']
})
data_frame
```

```python
data_frame.to_csv('pandas.csv') #to change it to csv file ,,, it will show the index number also
```

```python
data_frame.to_csv('pandas_index_false.csv',index = False)  #it  does not show the index number
```
